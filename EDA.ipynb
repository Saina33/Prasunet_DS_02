Filtering columns for analysis


```python
titanic_df = titanic_df[['Survived', 'Pclass', 'Sex', 'Age']]
titanic_df.head()
```

Transforming the data


```python
#0=Dead
#1=Survived
#pclass=1/1st,2/2nd,3/3rd
```

```python
titanic_df['Survived'] = titanic_df['Survived'].replace({0: 'Died', 1: 'Survived'})
titanic_df['Pclass'] = titanic_df['Pclass'].replace({1: '1st', 2: '2nd', 3: '3rd'})
titanic_df.head()
```

Relationships between variables


```python
#through cross tabulations
cross_tab = pd.crosstab(index=[titanic_df['Survived'], # First factor to compare
                               titanic_df['Sex']], # second factor to compare
                        columns=titanic_df['Pclass']) # the factor we want as our "column"
cross_tab

```

#Interpretation from cross tabulations
First Class had the highest survival rate for both men and women, with 89 female and 45 male survivors, compared to only 3 female and 77 male non-survivors.

Survival rates decrease and non-survivor numbers increase from first to third class, suggesting passenger class significantly impacted survival chances.

Most non-survivors were men in third class, with 300 men not surviving. Despite the high fatality rate in third class, the number of female survivors (72) equaled the number of non-survivors (72), indicating that the 'women and children first' policy might have been enforced during the evacuation.
